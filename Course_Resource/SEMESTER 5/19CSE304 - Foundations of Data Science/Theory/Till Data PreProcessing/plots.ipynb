```python
#Plotting multiple sets of data
plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.plot(x,y1, "b-" , label = '$y1 = x^2$') # Setting up legends 
plt.plot(x,y2, "r-" ,label ='$y2 = x^3$')   # Setting up legends
plt.xlabel("X - Axis",fontsize=20)
plt.ylabel("Y - Axis",rotation=45)
plt.title("this is my plot for power and cube")
plt.legend(loc='upper right')
#plt.show()
```

```python
plt.figure(figsize=(14,6))
x = np.linspace(0, 10, 100) 
y1 = np.sin(x) # Sine Graph
y2 = np.cos(x) # cosine graph
plt.subplot(1,2,1)  # 2,1,1
plt.plot(x,y1)
plt.subplot(1,2,2)  # 2,1,2
plt.plot(x,y2)
plt.show()
```

```python
plt.figure(figsize=(14,6))
x = np.linspace(0, 10, 100) 
y1 = np.sin(x) # Sine Graph
y2 = np.cos(x) # cosine graph
plt.subplot(2,1,1)  # 2,1,1
plt.plot(x,y1)
plt.subplot(2,1,2)  # 2,1,2
plt.plot(x,y2)
plt.show()
```

```python
x = np.arange(-50,50)
y1 = np.power(x,2)
y2 = np.power(x,3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)
```

```python
fig1 , ax1 = plt.subplots(nrows=3,ncols=3 , figsize = (20,20))
ax1[0,0].plot(x,y1)
ax1[0,0].set_title("power")
ax1[0,0].set_xlabel(r'$X$' , fontsize = 18) #Set the label for the x-axis
ax1[0,0].set_ylabel(r'$Y$' , fontsize = 18) #Set the label for the y-axis

ax1[0,1].plot(x,y2)

ax1[0,2].plot(x,y3)

ax1[1,0].plot(x,y4)

ax1[1,1].plot(x,y5)

ax1[1,2].plot(x,y6)

ax1[2,0].plot(x,y7)

ax1[2,1].plot(x,y8)

ax1[2,2].plot(x,y9)
```

```python
empl=pd.read_excel('/content/drive/MyDrive/FoDS-2021/data/visual/unemployment.xlsx')
empl.head()
```

```python
empl.dtypes
```

```python
# Show groups with different colors using "hue"
plt.figure(figsize=(14,7))

plt.style.use('fivethirtyeight')

# Group variable using "hue" that will produce lines with different colors
sns.lineplot(x="Period" , y="Unemployed" , hue="Gender" , data=empl,ci=None)   # ,ci=None
```

```python
plt.figure(figsize=(14,7))
sns.set(rc={'xtick.labelsize':17,'ytick.labelsize':10,'axes.labelsize':15})
sns.lineplot(x="Period" , y="Unemployed" , hue = "Gender" ,style="Gender" , markers=True , dashes=False ,data=empl,ci=None)
```

```python
# Combining lineplots using relplot
plt.figure(figsize=(10,10))
sns.set(rc={'xtick.labelsize':17,'ytick.labelsize':17,'axes.labelsize':20  , "axes.grid":False})
sns.relplot(x="Period" , y="Unemployed" , hue="Age" , col="Gender",kind='line', height=8.5, aspect=1 ,data=empl)
```

# Scatter Plot - matplotlib

```python
x1 = np.array([250,150,350,252,450,550,455,358,158,355])
y1 =np.array([40,50,80, 90, 100,50,60,88,54,45])


x2 = np.array([200,100,300,220,400,500,450,380,180,350])
y2 = np.array([400,500,800, 900, 1000,500,600,808,504,405])
```

```python
plt.scatter(x1,y1,color='r')
plt.scatter(x2,y2,marker='>')
plt.xlabel('$Time $ $ Spent$' , fontsize = 12)
plt.ylabel('$Score$' , fontsize = 12)
plt.title ('Scatter Graph')
```

```python
plt.figure(figsize=(8,6))
x = np.random.random(10)
y = np.random.random(10)
plt.scatter(np.random.random(10),np.random.random(10),c='r', s=50 , alpha=0.6 , label = 'One' )
plt.scatter(np.random.random(10),np.random.random(10),c='b', s=100 , alpha=0.6 , label = 'Two')
plt.scatter(np.random.random(10),np.random.random(10),c='g', s=150 , alpha=1 , label = 'Three')
plt.scatter(np.random.random(10),np.random.random(10),c='y', s=200 , alpha=0.6 , label = 'Four')
plt.legend()
```
